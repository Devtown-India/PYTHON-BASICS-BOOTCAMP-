###**What is OpenCV**
OpenCV stands for **Open Source Computer Vision Library**. As the name suggest it is a large open source library for computer vision, machine learnig, and image processing. It is a library that provides us with a lot of in-built function for image processing and computer vision that makes our code shorter and makes our work easier. The library has more than 2500 optimized algorithms, which includes a classic and state-of-art computer vision and machine learning algorithms. These algorithms can be used to detect and recognize faces, identify objects, classify human actions in videos, track camera movements, track moving objects, extract 3D models of objects, produce 3D point clouds from stereo cameras, stitch images together to produce a high resolution image of an entire scene, find similar images from an image database, remove red eyes from images taken using flash, follow eye movements, recognize scenery and establish markers to overlay it with augmented reality, etc.
To know more about OpenCV you can refer the [official documentation](https://opencv.org/about/).

###**Reading image using cv2**

For reading any image using OpenCv, we need **imread** function of cv2
<br>The imread function takes two parameters as the input named as *path* and *flag*.
- The first parameter *path* gives the location of the image to be read.
- The second parameter specifies the way image is to be read.
 - There are a lot of ways to read the image. A few of them are: 
   - **cv2.IMREAD_COLOR**: We can use cv2.IMREAD_COLOR or number 1 to specify this flag. This specifies to read the image as a three channel image (i.e. RGB image). Any alpha/tranparency channel will be neglected.
   - **cv2.IMREAD_GRAYSCALE**: It specifies to load the image in *grayscale* mode. We can either use number '0' or cv2.IMREAD_GRAYSCALE to specify this flag.
   - There are lot of other flags as well, to know about them in detail, you can reffer the [official documentation](https://docs.opencv.org/4.5.2/d8/d6a/group__imgcodecs__flags.html) 

<br>

> **NOTE:** The default value of the flag is cv2.IMREAD_COLOR

We can see the image is stored as a 3-D Array, with where first 2 dimensions defines the length and breadth of the image, while the 3rd dimension represents the three channel of the image (Red, Green and Blue).

###**Changing the image type**
We can convert any image from one form to another using **cvtColor** function of cv2.
<br>
> The function takes two inputs as parameter:
 - The first one is the variable storing the image.
 - The second parameter defines the conversion of the image from current form to the new one.
 - There are a lot of conversion flags available, you can refer those from the [official documentation](https://docs.opencv.org/3.4/d8/d01/group__imgproc__color__conversions.html).

```python
img = cv2.imread(img_path)

# Converts the RGB image to BGR image
im_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
print("RGB IMAGE")
cv2_imshow(img)
print("\nBGR IMAGE")
cv2_imshow(im_bgr)
```
